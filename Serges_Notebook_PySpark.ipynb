```
from pyspark.sql.functions import avg

data_df = spark.createDataFrame([("Broke", 20),("Denny", 31), ("Jules", 30), ("TD", 35), ("Broke", 25)], ["name", "age"])
#group thesame names together, aggregate their ages, and compute an average
avg_df = data_df.groupBy("name").agg(avg("age"))

#show the results of the final execution
display(avg_df)
```

```
#on importe la sparkSession et PySpark
import sys
from pyspark.sql import SparkSession
from pyspark.sql.functions import count

if __name__ == "__main__":
  if len(sys.argv) != 2:
    print("usage: mnmcount <file>", file=sys.stderr)
    sys.exit(-1)
    
#On cree la spark session
spark = (SparkSession
        .builder
        .appName("PythonMnMCount")
        .getOrCreate())
#récupération du fichier a partir de l'argument de ligne de commande
mnm_file = sys.argv[1]

#on lit le fichier dans un DataFrame
mnm_df = (spark.read.format("csv")
         .option("header", "true")
         .option("inferShema", "true")
         .load(mnm_file))

#on utilise le DataFram
count_mnm_df = (mnm_df
               .select("State", "Color", "Count")
               .groupBy("State", "Color")
               .agg(count("Count").alias("Total"))
               .orderBy("total", ascending=False))

#on affiche les résultats pour les colone etat et couleur
count_mnm_df.display(n=60, truncate=False)
print("Total Rows = %d" % (count_mnm_df_count()))

ca_count_mnm_df = (mnm_df
                  .select("State", "Color", "Count")
                  .where(mnm_df.State == "CA")
                  .groupBy("State", "Color")
                  .agg(count("Count").alias("Total"))
                  .orderBy("Total", ascending=False))

display(ca_count_mnm_df)
ca_count_mnm_df
spark.stop()
```
